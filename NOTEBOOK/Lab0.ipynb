#### The iterrows() Method

Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers. (Here, use the iterrows method to iterate each row of the dataframe and copy it to a dictionary. However, there is a easier way. Learn how the to_dict() method works by yourself later)

````python
#Get the column names of the Dataframe in order to make keys.
list_column_names = list(Salaries_df.reset_index().columns)

#create a dictionary
dictionary_salary = {}

#No.of columns in the dataframe
len_list = len(list_column_names)

#adding the colun names as keys, with an empty list assigned to each key
for i in range(len_list):
    dictionary_salary[list_column_names[i]] = []
    
#using the iterrows function to add the data to the dictionary    
for i in Salaries_df.iterrows():
    dictionary_salary[list_column_names[0]].append(i[0]) #adding the index column
    for j in range(len_list-1):
        dictionary_salary[list_column_names[j+1]].append(i[1][j]) #appending data of the columns other than the index column


        
        
    
    
    
````

#### The Slice Operation of ndarray Objects

Understand how the slice operation works for 1-D array and 2-D array.

Example:

```python
>>> arr = np.array([[1, 2, 3], [3, 4, 6], [7, 8, 9]])
>>> arr[1:]
array([[3, 4, 6],
       [7, 8, 9]])
>>> arr[1:, 0:2]
array([[3, 4],
       [7, 8]])


````python
#1-D Array

arr = np.array([1, 2, 3, 3, 4, 6, 7, 8, 9])
print(arr[2:,])
print(arr[1])

#2-D Array
print(np_list1[0:,])
print(np_list1[2:3,1:2])
print(np_list1[[1,1],[1,1]])
````

In a 1-D array arr, arr[x:y,] gives all the data in the array, from starting point x to ending point y. arr[x] gives thedata point at index x.

In a 2-D Array, arr[x:y,m:n] gives all the data in the intersection of rows x to y excluding y and columns m to n excluding n. 

In a 2-D Array, list can also be used as in index to reference a specific data point.
